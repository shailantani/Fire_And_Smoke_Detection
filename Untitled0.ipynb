```python
import os

# Define the path to your data.yaml file
data_yaml_path = '/content/drive/MyDrive/Fire_And_Smoke_detection/data/data.yaml'

print(f"Current working directory: {os.getcwd()}")
print(f"Verifying existence of data.yaml at: {data_yaml_path}")

# Check if the data.yaml file exists before attempting to open it
if not os.path.exists(data_yaml_path):
    print(f"Error: data.yaml not found at {data_yaml_path}. Please double check the path and ensure Google Drive is properly mounted and accessible.")
else:
    print(f"data.yaml found at {data_yaml_path}. Proceeding with modification.")
    # Read the current content of data.yaml
    with open(data_yaml_path, 'r') as f:
        lines = f.readlines()

    new_lines = []
    # Flag to check if 'path:' line has been processed
    path_line_processed = False

    for line in lines:
        stripped_line = line.strip()
        if stripped_line.startswith('path:'):
            # Update the 'path' to the absolute directory where 'train', 'val', 'test' are located
            new_lines.append(f"path: /content/drive/MyDrive/Fire_And_Smoke_detection/data\n")
            path_line_processed = True
        elif stripped_line.startswith('train:'):
            new_lines.append('train: train/images\n') # Relative to the 'path' above
        elif stripped_line.startswith('val:'):
            new_lines.append('val: val/images\n')     # Relative to the 'path' above
        elif stripped_line.startswith('test:'):
            new_lines.append('test: test/images\n')   # Relative to the 'path' above
        else:
            new_lines.append(line)

    # If 'path:' line was not found, add it at the beginning (or where appropriate)
    if not path_line_processed:
        new_lines.insert(0, f"path: /content/drive/MyDrive/Fire_And_Smoke_detection/data\n")

    # Write the modified content back to data.yaml
    with open(data_yaml_path, 'w') as f:
        f.writelines(new_lines)

    print(f"Successfully updated {data_yaml_path} with corrected paths.")

    # Display the updated content for verification
    print('\n--- Contents of updated data.yaml ---\n')
    with open(data_yaml_path, 'r') as f:
        print(f.read())
```
